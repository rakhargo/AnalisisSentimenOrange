```python
import pandas as pd
import re
```

```python
# Membaca data dari file CSV
df = pd.read_csv('data_review.csv')
df
```

```python
def clean_review(row):
    nama_pembeli = str(row['Nama pembeli']) if pd.notna(row['Nama pembeli']) else ''
    column4_text = str(row['Column 4'])
    tanggal_pembelian = str(row['Tanggal pembelian']) if pd.notna(row['Tanggal pembelian']) else ''

    if nama_pembeli:
        column4_text = column4_text.replace(nama_pembeli, "")
    
    if pd.isna(row['Nama pembeli']):
        column4_text = column4_text[7:]
    
    # Jika ada tanggal pembelian, hapus tanggal beserta teks sebelumnya
    if tanggal_pembelian:
        column4_text = column4_text.replace(tanggal_pembelian, "")

    # Hapus informasi metadata lainnya
    column4_text = re.sub(r'Laporkan Penyalahgunaan', '', column4_text, flags=re.IGNORECASE)
    column4_text = re.sub(r'Membantu\?', '', column4_text)
    column4_text = re.sub(r'\d$', '', column4_text)

    # column4_text = re.sub(r'\s+', ' ', column4_text).strip()
    
    return column4_text
```

```python
df['Review'] = df.apply(clean_review, axis=1)
df
```

```python
rating_cycle = [5, 4, 3, 2, 1]
rating_length = len(rating_cycle)
ratings = []

for i in range(len(df)):
    ratings.append(rating_cycle[(i // 66) % rating_length])

df['Rating'] = ratings
df['Sentimen'] = df['Rating'].apply(lambda x: 'Positif' if x > 3 else 'Negatif')
df
```

```python
df = df[df['Review'] != '']
pd.set_option('display.max_rows', None)  # Menampilkan semua baris
pd.set_option('display.max_columns', None)  # Menampilkan semua kolom
print(len(df))
df
```

```python
df = df.applymap(lambda x: x.replace('\n', ' ') if isinstance(x, str) else x)
print(len(df))
```

```python
df.to_csv('cleaned_data_review.csv', index=False)
```

```python
import torch
from transformers import BertTokenizer, BertForSequenceClassification
from torch import nn

model_name = "indolem/indobert-base-uncased"  # Ganti dengan model IndoBERT yang sesuai
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name)

# Kelas sentimen (Anda perlu menyesuaikan label jika model menggunakan label yang berbeda)
sentiments = ["Negatif", "Netral", "Positif"]

# Fungsi untuk menganalisis sentimen
def analyze_sentiment(text):
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512)
    
    # Prediksi menggunakan model
    with torch.no_grad():
        logits = model(**inputs).logits

    # Menghitung probabilitas dan menentukan kelas sentimen
    probabilities = torch.softmax(logits, dim=-1)
    predicted_class = torch.argmax(probabilities, dim=-1).item()
    
    return sentiments[predicted_class], probabilities[0][predicted_class].item()

# Menganalisis setiap review
results = []
for review in review:
    sentiment, probability = analyze_sentiment(review)
    results.append((review, sentiment, probability))

# Menampilkan hasil
for review, sentiment, probability in results:
    print(f"Review: {review}\nSentimen: {sentiment} (Probabilitas: {probability:.4f})\n")
```
